# Euler 4

A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 * 99.

Find the largest palindrome made from the product of two 3-digit numbers.


```j
threeDigitNumbers =: 100 + i. 900
pretty =: <"0 @(<./ , >./)
pretty threeDigitNumbers
```

```j
products =: ~. @, @(*/ ~) threeDigitNumbers
pretty products
```

```j
reverse =: |. @ ":
reverse 10
```

```j
isPalindrome =: (reverse -: ":)"0
isPalindrome 10 11 121 131 9009
```

```j
filterPalindromes =: (isPalindrome # ]) 
filterPalindromes 10 11 12 13 121
```

```j
>./ filterPalindromes products
```

```j
>./ ((((|. @ ":) -: ":)"0 # ]) ~. @, @(*/ ~) 100 + i. 900)
```
